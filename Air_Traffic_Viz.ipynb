```python
del_data = []
del_data = df_dom_2016['DEL-2016'].to_list() + df_dom_2017['DEL-2017'].to_list() + df_dom_2018['DEL-2018'].to_list() + df_dom_2019['DEL-2019'].tolist()
months = ['jan 2016', 'feb 2016', 'mar 2016', 'apr 2016', 'may 2016', 'jun 2016', 
         'jul 2016', 'aug 2016', 'sep 2016', 'oct 2016', 'nov 2016', 'dec 2016',
          'jan 2017', 'feb 2017', 'mar 2017', 'apr 2017', 'may 2017', 'jun 2017', 
         'jul 2017', 'aug 2017', 'sep 2017', 'oct 2017', 'nov 2017', 'dec 2017',
          'jan 2018', 'feb 2018', 'mar 2018', 'apr 2018', 'may 2018', 'jun 2018', 
         'jul 2018', 'aug 2018', 'sep 2018', 'oct 2018', 'nov 2018', 'dec 2018',
          'jan 2019', 'feb 2019', 'mar 2019', 'apr 2019', 'may 2019', 'jun 2019', 
         'jul 2019', 'aug 2019', 'sep 2019', 'oct 2019', 'nov 2019', 'dec 2019']
plt.figure(figsize=(50,10))
plt.plot(months, del_data)
plt.show
```

```python
def get_mae(max_leaf_nodes, train_X, val_X, train_y, val_y):
    model = DecisionTreeRegressor(max_leaf_nodes=max_leaf_nodes, random_state=0)
    model.fit(train_X, train_y)
    preds_val = model.predict(val_X)
    mae = mean_absolute_error(val_y, preds_val)
    return(mae)
y = del_data
x = pd.DataFrame([i for i in range(48)])
train_X, val_X, train_y, val_y = train_test_split(x, y, random_state=1)
del_model = DecisionTreeRegressor(random_state=1)
del_model.fit(train_X,train_y)
predictions = del_model.predict(val_X)
print(predictions)

#plt.figure(figsize=(50,10))
#plt.plot(val_X[0],predictions)
#plt.show
```

```python
candidate_max_leaf_nodes = [5, 25, 50, 100, 250, 500]
scores = {leaf_size: get_mae(leaf_size, train_X, val_X, train_y, val_y) for leaf_size in candidate_max_leaf_nodes}
best_tree_size = min(scores, key=scores.get)
```

```python
final_model = DecisionTreeRegressor(max_leaf_nodes=best_tree_size, random_state=1)
final_model.fit(x,y)
```

```python
predictions_final = final_model.predict(x)
print(predictions_final)
plt.figure(figsize=(50,10))
plt.plot(months, predictions_final)
plt.show
```
